```python
import matplotlib.pyplot as plt
import h5py
import numpy as np
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes, mark_inset
from matplotlib.ticker import StrMethodFormatter
import os
plt.rcParams.update({'font.size': 14})
```

```python
algorithms = ["FedRob","FedAvgAT","FedAvg"]
linestyles = ['-', '-', '-', '-',':', '--','-.']
markers = ["d","p","v","s","x","*","o"]
colors = [ 'tab:green', 'r', 'darkorange', 'tab:blue','tab:brown', 'm','slategray']

x = [0.25,0.5,0.75,1]
xi = list(range(len(x)))
y= [[0.74,0.71,0.48,0.30],[0.74,0.69,0.45,0.24],[0.62,0.42,0.25,0.15]]
for i in range(3):
    plt.plot(xi, y[i], linestyle=linestyles[i], label=algorithms[i], linewidth  = 1, color=colors[i], marker = markers[i],markevery=0.2, markersize=5)
plt.xticks(xi, x)
plt.ylabel('Robust Accuracy')
plt.xlabel('Percentage of attacked clients')
plt.title("PGD attack on MNIST")
plt.ylim([0.2,  0.8])
plt.legend(loc='lower left', prop={'size': 12}, ncol=3)
plt.grid(True)
plt.savefig("MNIST_PGD.pdf", bbox_inches="tight",pad_inches = 0)
plt.show()
```
