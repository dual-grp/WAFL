```python
import matplotlib.pyplot as plt
import h5py
import numpy as np
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes, mark_inset
from matplotlib.ticker import StrMethodFormatter
import os
plt.rcParams.update({'font.size': 14})
```

```python
algorithms = ["FedRob","FedPGD","FedFGSM", "FedAvg"]
linestyles = ['-', '-', '-', '-',':', '--','-.']
markers = ["*","p","v","s","x","d","o"]
colors = [ 'tab:green', 'r', 'darkorange', 'tab:blue','tab:brown', 'm','slategray']

x = [0, 0.2, 0.4, 0.6, 0.8]
xi = list(range(len(x)))
y= [[0.81,0.73,0.67,0.59, 0.51], [0.85,0.74,0.64,0.55, 0.44],[0.82,0.73,0.64, 0.56, 0.43], [0.87,0.70,0.54,0.32, 0.18]]
for i in range(len(y)):
    plt.plot(xi, y[i], linestyle=linestyles[i], label=algorithms[i], linewidth  = 1, color=colors[i], marker = markers[i],markevery=0.1, markersize=5)
plt.xticks(xi, x)
plt.ylabel('Robust Accuracy')
plt.xlabel('Percentage of attacked clients')
plt.title("PGD attack on MNIST")
plt.ylim([0.4,  0.9])
plt.legend(loc='lower left', prop={'size': 12}, ncol=1)
plt.grid(True)
plt.savefig("MNIST_PGD.pdf", bbox_inches="tight",pad_inches = 0)
plt.show()
```
